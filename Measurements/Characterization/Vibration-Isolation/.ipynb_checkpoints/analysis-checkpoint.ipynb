```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("axial-setup1.csv")
df2 = pd.read_csv("axial-setup2.csv")
df3 = pd.read_csv("axial-setup3.csv")

# Find rough resonance frequencies
res1 = df1["f"].iloc[df1["dB"].argmax()]
res2 = df2["f"].iloc[df2["dB"].argmax()]
res3 = df3["f"].iloc[df3["dB"].argmax()]

plt.figure(figsize=(10,6))
plt.plot(df1["f"], df1["dB"], label="5 Plate", marker=".")
plt.plot(df3["f"], df3["dB"], label="5 Plate Reverse", marker=".")
plt.plot(df2["f"], df2["dB"], label="3 Plate", marker=".")
plt.legend()
plt.xscale("log")
plt.grid(True, which="major", c="k")
plt.grid(True, which="minor", c="gray", linestyle="--")
plt.xlabel("$f$ / Hz")
plt.ylabel("Vibration transfer ratio / dB")

plt.savefig("axial.png")
```
